```python
df_imagenet_dogs=df_imagenet.loc[df_imagenet.wordnet_id.isin(df_dogs_imagenet.ImageNet_Synset.values)]
df_imagenet_dogs.reset_index(inplace=True,drop=True)
df_imagenet_dogs.shape
```

```python
df_imagenet_dogs['dog_group']='Unknown'
count=0
from tqdm.notebook import tqdm

for i in tqdm(range(120)):
    fuzzy_match=process.extractOne(df_imagenet_dogs.label.values[i],df_dog_groups.breed.values)
    if(fuzzy_match[1]>80):
        df_imagenet_dogs.loc[i,'dog_group']=df_dog_groups.loc[df_dog_groups.breed == fuzzy_match[0],'dog_group'].values
        count+=1
count
```

```python
df_imagenet_dogs['gender_bias']='Women-majority'
df_imagenet_dogs.loc[df_imagenet_dogs.mean_gender_audit>0.5,'gender_bias']='Men-majority'
df_imagenet_dogs.gender_bias.value_counts()
```

```python
chart = sns.violinplot(x="dog_group", y="mean_gender_audit", hue="gender_bias",
                    data=df_imagenet_dogs, palette="Set2", split=True,
                    scale="count", inner="stick")
chart.set_xticklabels(
    chart.get_xticklabels(), 
    rotation=45, 
    horizontalalignment='right',
    fontweight='light');
chart.plot(0.5*np.ones(8),'r-.')
plt.savefig('./plots_paper/dog_gender_violin.pdf',bbox_inches='tight')
```

# Musical instruments bias?

```python
list_music=['banjo'
'accordion',
'piano',
'chime',
'maraca',
'drum',
'banjo',
'cello',
'violin',
'harp',
'guitar',
'trumpet',
'french horn',
'trombone',
'harmonica',
'flute',
'oboe',
'saxophone']
```

```python
N_music=len(list_music)
```

```python
df_imagenet.shape
```

```python
count=0
music_imagenet_list=[]
for i in tqdm(range(N_music)):
    fuzzy_match=process.extractOne(list_music[i],df_imagenet.label.values)
    if(fuzzy_match[1]>80):
        print(fuzzy_match)
        music_imagenet_list.append(fuzzy_match[0])
        count+=1
count,len(music_imagenet_list)
```

```python
df_imagenet_music=df_imagenet.loc[df_imagenet.label.isin(music_imagenet_list)]
df_imagenet_music.reset_index(inplace=True,drop=True)
df_imagenet_music['gender_bias']='Women-majority'
df_imagenet_music.loc[df_imagenet_music.mean_gender_audit>0.5,'gender_bias']='Men-majority'
df_imagenet_music.gender_bias.value_counts()
```

```python
df_imagenet_music.head()
```

```python
df_srt=df_imagenet_music.sort_values('mean_gender_audit')
```

Source: http://www-usr.rider.edu/~vrme/v4n1/visions/Walker%20Influences%20of%20Gender%20and%20Sex

Abeles & Porter undertook the first major study of sex stereotyping of musical instruments
in 1978. They wrote that children found the drums to be the most masculine and the flute to
be the most feminine. 
Griswold and Chroback (1981) found in their study that:
answers by adults to hypothetical selection suggest [of instruments for their children to
play] that the sex of the child, rather than the sex of the adult or past musical experience,
influenced the adult's preferences. Violin, flute and clarinet were considered feminine;
drums, trombone, and trumpet were considered masculine; saxophone and cello were rated
neutral. (p. 57)
